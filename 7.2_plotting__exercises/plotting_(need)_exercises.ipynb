```python

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr= x_err, yerr= y_err,
             color= 'brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',                  # Marker o gives the circular marker on points, yellow gives marker colour
             ecolor='black',capsize=5,                              # Ecolor is error colour, capsize is size of the caps on errors
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')                      # Where the legend is
plt.show()

```

```python
x_data = np.linspace(0,50,10000)                # Bruh linspace gives u the points for free

y_data = (np.e)**(-x_data/10)*np.sin(x_data)    # Carries out the math on x points to get y

plt.plot(x_data,y_data,c='tab:blue')            # Line stuff
plt.ylim([-1, 1])                               # y axis limits

plt.xlabel('x',c='orange', fontsize = 15)
plt.ylabel('y',c='orange',fontsize = 15, rotation = 360)    # Why do I need to rotate it by 360 degrees for it to be vertical??
plt.title('$y=e^{-x/10}sin(x)$', fontsize = 20)
plt.grid(alpha=.25)
```
