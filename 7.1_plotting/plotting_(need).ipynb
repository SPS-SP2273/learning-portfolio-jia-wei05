### Things to note

```python

from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='I am legend',                      # Provides data as input for x and y axes, label gives the label for legend
         color='red', linestyle='dashed')                  # Customises plotted line
plt.xlabel('x-values')
plt.ylabel('y-values')                                     # Labels for x and y axes
plt.title('X vs Y')                                        # Plot title
plt.grid(alpha=.9)                                         # Contrast of grid lines
plt.legend(loc='center')                                   # Location of legend
```

## 4.1 Plotting data from files

```python
import numpy as np
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-5, 5])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```
