# Task 1

```python
import glob
import os
import shutil
```

```python
month_dict = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

file_name = 'oct-08_singapore.txt'
month_key = file_name.split('-')[0]                             # Gets the string before the dash (month)
new_name = file_name.replace(month_key, month_dict[month_key])  # Takes the string before the dash, use it as a key in the dict, replaces it
print(new_name)
```

# Task 6

```python
cairo_list = glob.glob('*cairo*') # NOTE: If search term is in the middle, need to be flanked by wildcards (*).
print(cairo_list)
```

```python
for file in cairo_list:
    path_to_destination = os.path.join('Locations','Cairo')
    shutil.copy(file, path_to_destination)                      # Copy file to path
    print(f'Copied {file} to {path_to_destination}')
```

# Task 7

```python
locations_remaining = [ 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

# First layer of loop to create list of text files for country
for location in locations_remaining:
    file_list7 = glob.glob(os.path.join('*' + location + '*',))     # Creates list of all files with specific country in the name by joining strings together with variable
    print('*' + location + '*')

    # Second layer of loop to transfer all text files into location file
    for file in file_list7:
        path_to_destination = os.path.join('Locations', location)       # Creates path by joining overarching Locations file with country name
        shutil.copy(file, path_to_destination)
        print(f'Copied {file} to {path_to_destination}')
```
